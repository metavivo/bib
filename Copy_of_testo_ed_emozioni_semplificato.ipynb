```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
from nltk.tokenize import RegexpTokenizer
from google.colab import drive
import csv


global word_tokenizer
global data_path
global data_filename
global data
global emotion_tags


def car(inlist):
    if inlist == []:
        return inlist
    else:
        return inlist[0]

def cdr(inlist):
    if inlist == []:
        return inlist
    else:
        return inlist[1:len(inlist)]

def cons(el, inlist):
    outlist = append([], inlist)
    outlist.insert(0, el)
    return outlist

def list(el):
    return cons(el, [])

def append(list1, list2):
    outlist = []
    for el in list1:
        outlist.append(el)
    for el in list2:
        outlist.append(el)
    return outlist

def filter(op, inlist):
    l = inlist
    accum = []
    if inlist == []:
        return inlist
    else:
        while len(l) > 0:
            if op(l[0]):
                accum = append(accum, list(car(l)))
            l = cdr(l)
    return accum

def load_data():
    return load_data1(data_filename)

def load_data1(data_filename):
    with open(data_filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        out = {}
        for line in csv_reader:
            if line[0] != 'NO TRANSLATION' and line[0] != '\ufeffItalian (it)':
                out[line[0]] = [int(x) for x in line[1:]]
        return out

def word_to_info(word):
    return word_to_info1(word, data)

def word_to_info1(word, data):
    if word in data:
        return data[word]

def word_to_emotions(word):
    info = word_to_info(word)
    if info != None:
        emo_info_list = info[2:]
        emo_tag_list = emotion_tags[2:]
        i = 0
        pairs = []
        while i < 8:
            pairs = pairs + [[emo_tag_list[i], emo_info_list[i]]]
            i = i + 1
        filtered_pairs = filter(lambda x: x[1] == 1, pairs)
        out = [x[0] for x in filtered_pairs]
        if out != []:
            return out

def text_to_emotions_info(input_text):
    step1 = word_tokenizer.tokenize(input_text)
    step2 = [[x, word_to_emotions(x)] for x in step1]
    step3 = filter(lambda x: None != x[1], step2)
    return step3

def get_emotion_text(text):
  step1 = text_to_emotions_info(text)
  step2 = [x[1] for x in step1]
  step3 = [' '.join(x) for x in step2]
  step4 = ' '.join(step3)
  return step4

def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off");

import shutil
shutil.rmtree('/content/datasetAndLibraryToInstall', ignore_errors=True)

!git clone https://github.com/matte156/datasetAndLibraryToInstall.git

data_path = '/content/datasetAndLibraryToInstall/step3 - prime emozioni e polaritÃ /'

data_filename = data_path + 'ita_emo.csv'

data = load_data()

word_tokenizer = RegexpTokenizer('\w+')

emotion_tags = ['Positive', 'Negative', 'rabbia', 'speranza', 'disgusto', 
                'paura', 'gioia', 'tristezza', 'sorpresa', 'fiducia']
```
